```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = mnist.load_data()
'''
plt.imshow(x_train[0]) cada elemento é uma imagem, sendo essa simbolizada como uma lista 3 arrays, 
simbolizando as tonalidades de vermelho, verde e azul
plt.show()

print(x_train[0])
'''

#A primeira coisa que nós precisamos é saber o número de entrada, que no caso é a quantidade de pixels 
#Shape 1 é o número de colunas e o shape 2 é a linha do numpy
num_pix = x_train.shape[1] * x_test.shape[2]

#Primeiro vamos transformar tudo em um único dataset pegando a dimensão de cada pixel, no caso serão três(RGB), pelo número de pixels
x_train = x_train.reshape(x_train.shape[0], num_pix).astype('float')
x_test = x_test.reshape(x_test.shape[0], num_pix).astype('float')

#Vamos então bilinearizar os nossos números para usarmos 
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
#Outra coisa que precisamos são as saídas, basicamente as 9 classes possíveis
classes = y_train.shape[1]

model = Sequential()


#Vamos normalizar cada pixel
x_train = x_train / 255
x_test = x_test / 255


#Deixando essa função de ativação, não precisa normalizar(0~1)
model.add(Dense(120, input_dim = num_pix, activation = "relu"))
#Vamos fazer a camada de saída:
model.add(Dense(classes, activation = "softmax"))
model.compile(optimizer="sgd", loss = 'categorical_crossentropy', metrics=['acc'])
model.fit(x_train, y_train, epochs = 100, batch_size = 100)
```
